```python
import pandas as pd
import numpy as np

def get_rel_err(df, name1, name2):
    return np.sum(df[name1] - df[name2])/df.shape[0]

def get_abs_err(df, name1, name2):
    return np.sum(np.abs(df[name1] - df[name2]))/df.shape[0]

def get_std(df, name1, name2):
    return np.std(np.abs(df[name1] - df[name2]))

airports = ['YSSY', 'EGLL', 'ZBAA']
names = ['Pressure', 'Temperature', 'Rel Humidity', 'Wind Dir', 'Wind Speed']
variables = [('metar_press', 'gfs_press'), ('metar_temp', 'gfs_temp'), ('metar_rh', 'gfs_rh'), ('metar_wind_dir', 'gfs_wind_dir'), ('metar_wind_spd', 'gfs_wind_spd')]

for airport in airports:
    df = pd.read_hdf('/home/roz016/data.h5', airport)
    for i, name in enumerate(names):
        print(name)
        print("Abs Error: {}".format(get_abs_err(df, variables[i][0], variables[i][1])))
        print("Rel Error: {}".format(get_rel_err(df, variables[i][0], variables[i][1])))
        print("S: {}".format(get_std(df, variables[i][0], variables[i][1])))
    print('__________________')
```

```python
import math

a = [0,180]

def circular_variance(data):

    x = y = 0.0
    for value in data:
        x += math.cos(math.radians(value))
        y += math.sin(math.radians(value))

    return 1 - math.sqrt(math.pow(x/len(data), 2) + math.pow(y/len(data), 2))

circular_variance(a)
```

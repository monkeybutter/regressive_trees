```python
import pandas as pd
import seaborn as sb
import json

df_cx0 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx1.json')
df_cx1 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx2.json')
df_cx2 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx3.json')
df_cx3 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx4.json')
df_cx4 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx5.json')

df = pd.concat([df_cx0, df_cx1, df_cx2, df_cx3, df_cx4])

df['gfs_wspd_diff'] = df['gfs_wind_spd'] - df['metar_wind_spd']
df['tree_wspd_diff'] = df['single_tree'] - df['metar_wind_spd']
df['rf_wspd_diff'] = df['rf'] - df['metar_wind_spd']

sb.set(style="whitegrid")
sb.violinplot([df.gfs_wspd_diff, df.tree_wspd_diff, df.rf_wspd_diff], names=['GFS', 'Tree', 'RF'])
sb.despine(left=True)
```

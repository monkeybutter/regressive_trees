```python
import pandas as pd
import seaborn as sb
import json

df_cx0 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx1.json')
df_cx1 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx2.json')
df_cx2 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx3.json')
df_cx3 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx4.json')
df_cx4 = pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/yssy_metar_wind_spd_cx5.json')

df = pd.concat([df_cx0, df_cx1, df_cx2, df_cx3, df_cx4])

df['gfs_wspd_diff'] = df['gfs_wind_spd'] - df['metar_wind_spd']
df['tree_wspd_diff'] = df['single_tree'] - df['metar_wind_spd']
df['rf_wspd_diff'] = df['rf'] - df['metar_wind_spd']

print(df.gfs_wspd_diff.count())
print(df.gfs_wspd_diff.mean())
print(df.gfs_wspd_diff.std())
print(df.gfs_wspd_diff.min())
print(df.gfs_wspd_diff.max())
print(df.loc[df['gfs_wspd_diff'] < df.gfs_wspd_diff.mean()-1].gfs_wspd_diff.count())
print(df.loc[df['gfs_wspd_diff'] > df.gfs_wspd_diff.mean()+1].gfs_wspd_diff.count())
print(df.loc[df['gfs_wspd_diff'] < df.gfs_wspd_diff.mean()-2].gfs_wspd_diff.count())
print(df.loc[df['gfs_wspd_diff'] > df.gfs_wspd_diff.mean()+2].gfs_wspd_diff.count())

print(df.tree_wspd_diff.count())
print(df.tree_wspd_diff.mean())
print(df.tree_wspd_diff.std())
print(df.tree_wspd_diff.min())
print(df.tree_wspd_diff.max())
print(df.loc[df['tree_wspd_diff'] < df.tree_wspd_diff.mean()-1].tree_wspd_diff.count())
print(df.loc[df['tree_wspd_diff'] > df.tree_wspd_diff.mean()+1].tree_wspd_diff.count())
print(df.loc[df['tree_wspd_diff'] < df.tree_wspd_diff.mean()-2].tree_wspd_diff.count())
print(df.loc[df['tree_wspd_diff'] > df.tree_wspd_diff.mean()+2].tree_wspd_diff.count())

print(df.rf_wspd_diff.count())
print(df.rf_wspd_diff.mean())
print(df.rf_wspd_diff.std())
print(df.rf_wspd_diff.min())
print(df.rf_wspd_diff.max())
print(df.loc[df['rf_wspd_diff'] < df.rf_wspd_diff.mean()-1].rf_wspd_diff.count())
print(df.loc[df['rf_wspd_diff'] > df.rf_wspd_diff.mean()+1].rf_wspd_diff.count())
print(df.loc[df['rf_wspd_diff'] < df.rf_wspd_diff.mean()-2].rf_wspd_diff.count())
print(df.loc[df['rf_wspd_diff'] > df.rf_wspd_diff.mean()+2].rf_wspd_diff.count())

sb.set(style="whitegrid")
sb.violinplot([df.gfs_wspd_diff, df.tree_wspd_diff, df.rf_wspd_diff], names=['GFS', 'Tree', 'RF'])
sb.despine(left=True)
```

```python
import numpy as np
import pandas as pd

airports = ["yssy", "egll", "zbaa"]

for airport in airports:
    
    df = pd.read_csv("/home/roz016/Dropbox/Data for Tree/Results/new_tree_cx10_lin_vs_cir/wind_spd_{}.csv".format(airport))

    index = [2000, 1000, 500, 200, 100, 50]
    new_df = pd.DataFrame(index=index, columns=["linear_train", "circular_train", "linear_test", "circular_test"])
    new_df = new_df.fillna(0) # with 0s rather than NaNs

    for i in index:
        new_df.ix[i, 'linear_train'] = np.mean(df["lin_" + str(i) + "_cc_train"])
        new_df.ix[i, 'circular_train'] = np.mean(df["cir_" + str(i) + "_cc_train"])
        new_df.ix[i, 'linear_test'] = np.mean(df["lin_" + str(i) + "_cc_test"])
        new_df.ix[i, 'circular_test'] = np.mean(df["cir_" + str(i) + "_cc_test"])


    new_df.plot(title= airport, xticks=index, logx=True)
```

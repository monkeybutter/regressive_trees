```python
import pandas as pd
import seaborn as sb
import json

airports = ['yssy', 'egll', 'zbaa']
class_vars = ['metar_wind_spd', 'metar_press', 'metar_rh']

airport = airports[2]
class_var = class_vars[0]

dfs = []
for i in range(0, 5):
    dfs.append(pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_bin100/{}__regression_cx{}.json'.format(airport, i)))

df_reg = pd.concat(dfs)

df_reg['std_reg_diff'] = df_reg['std_regression'] - df_reg['metar_wind_spd']
df_reg['opt_reg_diff'] = df_reg['opt_regression'] - df_reg['metar_wind_spd']

dfs = []
for i in range(1, 6):
    dfs.append(pd.read_json('/home/roz016/Dropbox/Data for Tree/Results/cx5_rf50_bin100/{}_{}_cx{}.json'.format(airport, class_var, i)))

df_rf = pd.concat(dfs)

df_rf['gfs_wspd_diff'] = df_rf['gfs_wind_spd'] - df_rf['metar_wind_spd']
df_rf['tree_wspd_diff'] = df_rf['single_tree'] - df_rf['metar_wind_spd']
df_rf['rf_wspd_diff'] = df_rf['rf'] - df_rf['metar_wind_spd']


print("Count: {}, {}, {}, {}, {}".format(df_rf.gfs_wspd_diff.count(), 
                                         df_reg.std_reg_diff.count(), 
                                         df_reg.opt_reg_diff.count(),
                                         df_rf.tree_wspd_diff.count(),
                                         df_rf.rf_wspd_diff.count()))

print("Mean: {}, {}, {}, {}, {}".format(df_rf.gfs_wspd_diff.mean(), 
                                         df_reg.std_reg_diff.mean(), 
                                         df_reg.opt_reg_diff.mean(),
                                         df_rf.tree_wspd_diff.mean(),
                                         df_rf.rf_wspd_diff.mean()))

print("Std: {}, {}, {}, {}, {}".format(df_rf.gfs_wspd_diff.std(), 
                                         df_reg.std_reg_diff.std(), 
                                         df_reg.opt_reg_diff.std(),
                                         df_rf.tree_wspd_diff.std(),
                                         df_rf.rf_wspd_diff.std()))

print("Count Mean -1: {}, {}, {}, {}, {}".format(df_rf.loc[df_rf['gfs_wspd_diff'] < df_rf.gfs_wspd_diff.mean()-1].gfs_wspd_diff.count(), 
                                         df_reg.loc[df_reg['std_reg_diff'] < df_reg.std_reg_diff.mean()-1].std_reg_diff.count(),
                                         df_reg.loc[df_reg['opt_reg_diff'] < df_reg.opt_reg_diff.mean()-1].opt_reg_diff.count(),
                                         df_rf.loc[df_rf['tree_wspd_diff'] < df_rf.tree_wspd_diff.mean()-1].tree_wspd_diff.count(),
                                         df_rf.loc[df_rf['rf_wspd_diff'] < df_rf.rf_wspd_diff.mean()-1].rf_wspd_diff.count()))

print("Count Mean +1: {}, {}, {}, {}, {}".format(df_rf.loc[df_rf['gfs_wspd_diff'] > df_rf.gfs_wspd_diff.mean()+1].gfs_wspd_diff.count(), 
                                         df_reg.loc[df_reg['std_reg_diff'] > df_reg.std_reg_diff.mean()+1].std_reg_diff.count(),
                                         df_reg.loc[df_reg['opt_reg_diff'] > df_reg.opt_reg_diff.mean()+1].opt_reg_diff.count(),
                                         df_rf.loc[df_rf['tree_wspd_diff'] > df_rf.tree_wspd_diff.mean()+1].tree_wspd_diff.count(),
                                         df_rf.loc[df_rf['rf_wspd_diff'] > df_rf.rf_wspd_diff.mean()+1].rf_wspd_diff.count()))

print("Count Mean -2: {}, {}, {}, {}, {}".format(df_rf.loc[df_rf['gfs_wspd_diff'] < df_rf.gfs_wspd_diff.mean()-2].gfs_wspd_diff.count(), 
                                         df_reg.loc[df_reg['std_reg_diff'] < df_reg.std_reg_diff.mean()-2].std_reg_diff.count(),
                                         df_reg.loc[df_reg['opt_reg_diff'] < df_reg.opt_reg_diff.mean()-2].opt_reg_diff.count(),
                                         df_rf.loc[df_rf['tree_wspd_diff'] < df_rf.tree_wspd_diff.mean()-2].tree_wspd_diff.count(),
                                         df_rf.loc[df_rf['rf_wspd_diff'] < df_rf.rf_wspd_diff.mean()-2].rf_wspd_diff.count()))

print("Count Mean +2: {}, {}, {}, {}, {}".format(df_rf.loc[df_rf['gfs_wspd_diff'] > df_rf.gfs_wspd_diff.mean()+2].gfs_wspd_diff.count(), 
                                         df_reg.loc[df_reg['std_reg_diff'] > df_reg.std_reg_diff.mean()+2].std_reg_diff.count(),
                                         df_reg.loc[df_reg['opt_reg_diff'] > df_reg.opt_reg_diff.mean()+2].opt_reg_diff.count(),
                                         df_rf.loc[df_rf['tree_wspd_diff'] > df_rf.tree_wspd_diff.mean()+2].tree_wspd_diff.count(),
                                         df_rf.loc[df_rf['rf_wspd_diff'] > df_rf.rf_wspd_diff.mean()+2].rf_wspd_diff.count()))


sb.set(style="whitegrid")
sb.violinplot([df_rf.gfs_wspd_diff, df_reg.std_reg_diff, df_reg.opt_reg_diff, df_rf.tree_wspd_diff, df_rf.rf_wspd_diff], names=['GFS', 'Std Reg', 'Opt Reg', 'Tree', 'RF'])
sb.axlabel("Methodology", "Error [knots]")
sb.despine(left=True)
```
